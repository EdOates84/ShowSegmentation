```python
import os
import face_recognition
import numpy as np
import pandas as pd 
from tqdm import tqdm_notebook as tqdm
import pickle
from os.path import join
root = '/home/eon/Desktop/face-clustering/dataset/'
output_path = 'test_encodings.pickle'
celebs = sorted(os.listdir(root))[:] #CHANGED
encodings = []

def load_pickle(filename):
    infile = open(filename,'rb')
    x = pickle.load(infile)
    infile.close()
    return x

def store_pickle(filename, data):
    file = open(filename,'wb')
    pickle.dump(data, file)
    file.close()

# On the cluster:
#   finish knowledge graph api fetching names
#   read the csv and rename all directories with their celeb name
#


for celeb in tqdm(celebs):
    #TODO: GET LARGEST IMAGES
    image_paths = os.listdir(os.path.join(root,celeb))[:5]
    encodings.append([])
    for path in image_paths:
        img = face_recognition.load_image_file(os.path.join(root, celeb, path))
        bboxes = face_recognition.face_locations(img,model='cnn') #TODO: Test if works well else take cnn
        enc = face_recognition.face_encodings(img, bboxes)
        encodings[-1].append(enc[0]) #0th value for 1st face - maybe more than 1 face?

encodings = np.array(encodings)
store_pickle(output_path, (celebs, encodings))

#TODO: KNN
# convert (celebs, encodings) data into the format required by KNN
# Fill KNN space with encodings of all celebs before looping over all faces
# def face2name(face_img):
    # get encodings of face_img - in case of multiple faces - take the largest face
    # pass it through KNN and infer
    # return the found name
```

```python
from sklearn.neighbors import KNeighborsClassifier
X = []
Y = []



for i in range(len(encodings)): #prepare dataset
    for encoding in encodings[i]:
        X.append(encoding)
        Y.append(i)
        
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X, Y)

for person in os.listdir('/home/eon/Desktop/face-clustering/dataset/'):
    for face in os.listdir('/home/eon/Desktop/face-clustering/dataset/{}'.format(person)):
        img = face_recognition.load_image_file('/home/eon/Desktop/face-clustering/dataset/{}/{}'.format(person,face))
        bboxes = face_recognition.face_locations(img,model='cnn') #TODO: Test if works well else take cnn
        enc = face_recognition.face_encodings(img, bboxes)
        try:
            print('{} {}'.format(person, neigh.predict(enc)))
        except:
            print(enc)
            print('/home/eon/Desktop/face-clustering/dataset/{}/{}'.format(person,face))
```

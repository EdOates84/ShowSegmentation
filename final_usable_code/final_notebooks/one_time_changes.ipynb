## Making a list of only news reporters

```python
celeb_info = pd.read_csv('celeb_info_ALL.csv')
synonyms = ['correspondent','host', 'anchor', 'news', 'broadcaster', 'reporter', 'journalist', 'personality', 'commentator']

og_len = len(synonyms)
for x in range(len(synonyms)):
    synonyms.append(synonyms[x].capitalize())
    
regex_query = '|'.join(synonyms)

#television personalit
```

```python
anchors = celeb_info[celeb_info['article_body'].str.contains(regex_query, na=False)]
anchors_with_television = celeb_info[celeb_info['article_body'].str.contains('Television|television|'+regex_query, na=False)]
```

```python
store_pickle('anchors-anchors_with_TV-tuple.pickle', (anchors, anchors_with_television))
```
